```python
import requests
import json
import time
import urllib3

# âœ… Suppress SSL warnings
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# âœ… Kafka REST Proxy Config
REST_PROXY_URL = "https://3efc35b4eaaec49b.p171.r3proxy.com"
TOPIC_NAME = "ddos-topic"
GROUP_NAME = "ddos-detector-group"
INSTANCE_ID = "colab-consumer"

# ===============================
# ğŸ”¹ Step 1: Create Consumer
# ===============================
def create_consumer():
    url = f"{REST_PROXY_URL}/consumers/{GROUP_NAME}"
    headers = {
        "Content-Type": "application/vnd.kafka.v2+json"
    }
    data = {
        "name": INSTANCE_ID,
        "format": "json",
        "auto.offset.reset": "earliest"
    }
    response = requests.post(url, headers=headers, data=json.dumps(data), verify=False)
    print(f"ğŸ“© Create Consumer Response: {response.status_code} - {response.text}")
    if response.status_code == 200:
        base_uri = response.json().get("base_uri")
        print(f"âœ… Consumer created. Base URI: {base_uri}")
        return base_uri
    else:
        raise Exception("âŒ Failed to create consumer")

# ===============================
# ğŸ”¹ Step 2: Subscribe to Topic
# ===============================
def subscribe_to_topic(consumer_instance_url):
    url = f"{consumer_instance_url}/subscription"
    headers = {
        "Content-Type": "application/vnd.kafka.v2+json"
    }
    data = {
        "topics": [TOPIC_NAME]
    }
    print(f"ğŸ“ Subscribing to: {url} with payload {data}")
    response = requests.post(url, headers=headers, data=json.dumps(data), verify=False)
    print(f"ğŸ“¤ Subscribe Response: {response.status_code} - {response.text}")
    if response.status_code != 204:
        raise Exception(f"âŒ Subscription failed: {response.text}")
    print(f"âœ… Subscribed to topic: {TOPIC_NAME}")

# ===============================
# ğŸ”¹ Step 3: Read Messages
# ===============================
def read_messages(consumer_instance_url):
    url = f"{consumer_instance_url}/records"
    headers = {
        "Accept": "application/vnd.kafka.json.v2+json"
    }
    print("â³ Polling messages...")
    response = requests.get(url, headers=headers, verify=False)
    print(f"ğŸ“¥ Poll Response: {response.status_code}")
    if response.status_code == 200:
        messages = response.json()
        for msg in messages:
            print(f"ğŸ“¦ Message: {msg['value']}")
        return messages
    elif response.status_code == 204:
        print("âš ï¸ No new messages.")
        return []
    else:
        print(f"âŒ Error while polling: {response.text}")
        return []

# ===============================
# ğŸ”¹ Step 4: Delete Consumer
# ===============================
def delete_consumer(consumer_instance_url):
    response = requests.delete(consumer_instance_url, headers={"Accept": "application/vnd.kafka.v2+json"}, verify=False)
    if response.status_code == 204:
        print(f"ğŸ§¹ Existing consumer deleted: {INSTANCE_ID}")
    else:
        print(f"âŒ Failed to delete consumer: {response.text}")

# ===============================
# ğŸ”¹ Full Flow Execution
# ===============================
try:
    consumer_instance_url = create_consumer()
    subscribe_to_topic(consumer_instance_url)
    time.sleep(1)  # Give time for subscription to settle
    messages = read_messages(consumer_instance_url)
finally:
    if 'consumer_instance_url' in locals():
        delete_consumer(consumer_instance_url)
```

```python
import requests
import json
import urllib3
import uuid

# âœ… Suppress SSL warnings for Remote.It self-signed certs
urllib3.disable_warnings(urllib3.exceptions.InsecureRequestWarning)

# âœ… Remote.It Kafka REST Proxy URL
REST_PROXY_URL = "https://11fada9e25e6be1e.p171.r3proxy.com"
GROUP_NAME = "auto-consumer-group"
INSTANCE_ID = f"consumer-{uuid.uuid4()}"

# âœ… Step 0: Get available topics
topics_url = f"{REST_PROXY_URL}/topics"
response = requests.get(topics_url, verify=False)
try:
    topics = response.json()
except Exception:
    raise Exception(f"âŒ Could not parse topics list:\n{response.text}")

print("âœ… Topics available:", topics)

if not topics or not isinstance(topics, list):
    raise Exception("âŒ No valid topic list returned from REST Proxy.")

# âœ… Select the last topic automatically
TOPIC_NAME = topics[-1].strip()
print(f"ğŸ“Œ Selected topic: {TOPIC_NAME}")

# âœ… Step 1: Produce a message to the topic
produce_url = f"{REST_PROXY_URL}/topics/{TOPIC_NAME}"
produce_payload = {
    "records": [{"value": {"user": "zabhiuddin...", "msg": f"Auto message to {TOPIC_NAME}"}}]
}
headers = {
    "Content-Type": "application/vnd.kafka.json.v2+json",
    "Accept": "application/vnd.kafka.v2+json"
}
response = requests.post(produce_url, headers=headers, data=json.dumps(produce_payload), verify=False)
print("ğŸ“¤ Produce response:", response.text)

# âœ… Step 2: Create a consumer
create_consumer_url = f"{REST_PROXY_URL}/consumers/{GROUP_NAME}"
consumer_payload = {
    "name": INSTANCE_ID,
    "format": "json",
    "auto.offset.reset": "earliest"
}
response = requests.post(create_consumer_url, headers={"Content-Type": "application/vnd.kafka.v2+json"}, data=json.dumps(consumer_payload), verify=False)
print("ğŸ§¾ Create consumer:", response.text)

# âœ… Step 3: Extract base_uri and enforce HTTPS
base_uri = response.json().get("base_uri")  # âœ… Correct: get the value of the 'base_uri' key
if not base_uri:
    raise Exception("âŒ base_uri not found in consumer creation response.")
base_uri = base_uri.replace("http://", "https://")
print(f"ğŸ”— Using base_uri: {base_uri}")

# âœ… Step 4: Subscribe to topic
subscription_url = f"{base_uri}/subscription"
subscription_payload = {"topics": [TOPIC_NAME]}
response = requests.post(subscription_url, headers={"Content-Type": "application/vnd.kafka.v2+json"}, data=json.dumps(subscription_payload), verify=False)

# âœ… Handle optional empty subscribe response
if response.status_code == 204 or response.text.strip() == "":
    print("ğŸ”” Subscribe response: âœ… Successfully subscribed (empty response)")
else:
    print("ğŸ”” Subscribe response:", response.text)

# âœ… Step 5: Fetch records
records_url = f"{base_uri}/records"
response = requests.get(records_url, headers={"Accept": "application/vnd.kafka.json.v2+json"}, verify=False)
print("ğŸ“¥ Fetched records:", response.text)
```
